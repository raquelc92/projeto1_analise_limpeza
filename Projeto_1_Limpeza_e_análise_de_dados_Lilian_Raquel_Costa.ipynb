**Passo 3**: Limpeza dos dados

Verificar a quantidade de dados faltantes no Dataset

```python
df.isnull().sum()
```

A coluna "Attribute" está com quase 100% sem dados, o que não traz qualquer relevância para nossa análise. Utilizaremos `del` para excluir essa coluna

```python
del df["attribute"]
```

```python
df.head()
```

```python
#Verificar dados duplicados no index

df[df.index.duplicated()]
```

```python
#Excluir dados duplicados
df = df[~df.index.duplicated()]
```

Coluna 'Player' - substituir todas as linhas que estão com '1 player' por 1

```python
print("Players column unique values:")
print(pd.Series(df.players.unique()), end='\n\n')
```

```python
df["players"].replace("1 Player", "1", inplace=True)
```

Na coluna 'platform', verificaremos as entradas e faremos a correção de eventuais itens que estejam repetidos, mas escritos de formas diferentes 	

```python
print("Platform column unique values:")
print(pd.Series(df.platform.unique()), end='\n\n')
```

```python
df["platform"].replace("pc", "PC", inplace=True)
df["platform"].replace("xBoX", "XBOX", inplace=True)
```

```python
#Renomear a coluna 'Genre(s), trocando para 'genre'
df.rename(columns={'genre(s)': 'genre'}, inplace = True)
```

```python
print("Genre column unique values:")
print(pd.Series(df.genre.unique()), end='\n\n')
```

```python
print("Rating column unique values:")
print(pd.Series(df.rating.unique()), end='\n\n')
```

```python
df["rating"].replace("aceitas pix", np.nan , inplace=True)
```

Podemos verificar também se há alguma relação importante entre os dados, utilizando a função "pairplot" do seaborn, na qual cria-se uma matriz de gráficos, em que a diagonal principal representa a distribuição em barras dos dados, e os demais representam o gráfico "scatter". Isso nos permite identificar se há alguma relação gritante nos nossos dados, além de identificar algum outlier em sua distribuição.

Faremos um pairplot das avaliações da crítica e das avaliações dos usuários.

A correlação informa como as variáveis ​​em um conjunto de dados estão relacionadas entre si e como elas se movem em relação umas às outras. O valor da correlação varia de -1 a +1. A correlação indica que duas variáveis ​​são independentes uma da outra. Uma correlação positiva indica que as variáveis ​​se movem na mesma direção e uma correlação negativa indica o oposto.

Aqui, utilizaremos um mapa de calor para representar os valores do array graficamente com diferentes tons de cores para diferentes valores.


```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.heatmap(df.corr(), vmin = -1, vmax = +1, annot = True, cmap = 'coolwarm')
```

```python

upp_mat = np.triu(df.corr())

sns.heatmap(df.corr(), vmin = -1, vmax = +1, annot = True, cmap = 'coolwarm', mask = upp_mat)
```

Podemos também verificar se existem possíveis correlações entre 'rating' e avaliações positivas e negativas da crítica. Para isso, usaremos 'Stripplots', gráficos que mostram como os dados são distribuídos por determinadas variáveis categóricas, demonstrando possíveis correlações.

```python
plt.figure(figsize=(15,10))
sns.stripplot(x="rating", y="critic_positive", data=df)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.stripplot(x="rating", y="critic_negative", data=df)
plt.show()
```
